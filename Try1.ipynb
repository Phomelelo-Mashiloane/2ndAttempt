```python
import pandas as pd
import numpy
import matplotlib as mp
```

```python
##read .csv file with the data

mobility = pd.read_csv('Global_Mobility_Report_062022.csv', dtype={'sub_region_2': 'str', 'metro_area': 'str', 'iso_3166_2_code': 'str'})
mobility.head()
```

```python
##Select only information for South Africa

SA = mobility.loc[mobility['country_region_code'] == 'ZA']
SA.tail()

SouthAfrica = SA.iloc[:,[2,9,10,11,12,13,14]]
#SouthAfrica = SA.iloc[:,[2,9:14]]
SouthAfrica.head()
```

```python
##Compare the average change over the period for all the regions.
##I want to get the averages separately
##then save the averages to a table

LP = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Limpopo']
MP = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Mpumalanga']
GP = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Gauteng']
NW = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'North West']
FS = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Free State']
NC = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Nothern Cape']
KZN = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'KwaZulu-Natal']
EC = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Eastern Cape']
WC = SouthAfrica.loc[SouthAfrica['sub_region_1'] == 'Western Cape']
KZN.head()
```

```python
#Get the averages
LA = LP['retail_and_recreation_percent_change_from_baseline'].mean()
MA = MP['retail_and_recreation_percent_change_from_baseline'].mean()
GA = GP['retail_and_recreation_percent_change_from_baseline'].mean()
NWA = NW['retail_and_recreation_percent_change_from_baseline'].mean()
FA = FS['retail_and_recreation_percent_change_from_baseline'].mean()
NCA = NC['retail_and_recreation_percent_change_from_baseline'].mean()
KA = KZN['retail_and_recreation_percent_change_from_baseline'].mean()
EA = EC['retail_and_recreation_percent_change_from_baseline'].mean()
WA = WC['retail_and_recreation_percent_change_from_baseline'].mean()
average = [LA, MA, GA, NWA, FA, NCA, KA, EA, WA]
print(average)
```
